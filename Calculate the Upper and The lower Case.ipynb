```python
def ULcalc(string):
    a = [chr(i) for i in range(97, 123)]
    A = [j.upper() for j in a]
    Dict = {"Number of Upper case Characters":0,"Number of Lower case Characters":0}    
    for ch in string:
        if ch in A:
            Dict["Number of Upper case Characters"] +=1
        elif ch in a:
            Dict["Number of Lower case Characters"] +=1
    return Dict
```
